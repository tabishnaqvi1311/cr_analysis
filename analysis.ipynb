### making changes

```python
#make changes to instance of dataframe 
df["Hit Speed"] = pd.to_numeric(df["Hit Speed"], errors="coerce")
df["Damage"] = pd.to_numeric(df["Damage"], errors="coerce")
df["DPS"] = pd.to_numeric(df["DPS"], errors="coerce")

df["rawDPS"] = df["Damage"]/df["Hit Speed"]
df["delta"] = abs(df["rawDPS"]-df["DPS"])

df.head(10)
```

### advanced filterng

```python
df.loc[(df["Rarity"] == "Legendary") | (df["Elixir Cost"] == "3")]
# 3 elixir legendary cards that do more than 100 dmg
df.loc[(df["Rarity"] == "Legendary") & (df["Elixir Cost"] == "3") & (df["Damage"] > 100)]
```

```python
#using literal strings
df.loc[df["Name"].str.contains("Goblin")]
#using regex
import re
df.loc[df["Name"].str.contains('fire|ice', flags=re.IGNORECASE,  regex=True)]
```

```python
wiz = df.loc[df["Name"].str.contains('[a-z]*wizard', flags=re.IGNORECASE, regex=True)]
wiz
```
